```python

# 1) OR gate*****************************************************************************************************************

def org(inp1,inp2):
    if (inp1==0):
        if(inp2==0):
            op=0
        else:
            op=inp2
    elif (inp2==0):
        if(inp1==0):
            op=0
        else:
            op=inp1    
    elif(inp1==1 or  inp2==1):
        op=1
    else:
        op=dc
    return op

# 2) AND gate***************************************************************************************************************

def andg(inp1,inp2):
    if (inp1==1):
        if(inp2==1):
            op=1
        else:
            op=inp2
    elif (inp2==1):
        if(inp1==1):
            op=1
        else:
            op=inp1    
    elif(inp1==0 or  inp2==0):
        op=0
    else:
        op=dc
    return op

# 3) NOT gate****************************************************************************************************************

def notg(inp):
    if(inp==1):
        op= 0
    elif(inp==0):
        op= 1
    else:
        op=inp+'bar'
    return op

# 4) XOR gate****************************************************************************************************************

def xorg(inp1,inp2):
    if(inp1==inp2):
        op=0
    else:
        if (inp1==1):
            if(inp2==0):
                op=1
            else:
                op=inp2+'_bar'
        elif (inp1==0):
            if(inp2==1):
                op=1
            else:
                op=inp2
        elif (inp2==0):
            if(inp1==1):
                op=1
            else:
                op=inp1   
        elif (inp2==1):
            if(inp1==0):
                op=1
            else:
                op=dc+'bar'
    return op

# 5) XNOR gate***************************************************************************************************************

def xnorg(inp1,inp2):
    if(inp1==inp2):
        op=1
    else:
        if (inp1==1):
            if(inp2==0):
                op=0
            else:
                op=inp2
        elif (inp1==0):
            if(inp2==1):
                op=0
            else:
                op=inp2+'bar'
        elif (inp2==1):
            if(inp1==0):
                op=0
            else:
                op=inp1    
        elif (inp2==0):
            if(inp1==1):
                op=0
            else:
                op=inp1+'bar' 
    return op
```

```python
# Global backtracking functions

def and_bt(op,inp1,inp2):
    if(op==1 or op=='d'):
        inp1=1
        inp2=1
    elif(op==0):
        if(inp1==1):
            inp2=0
        elif(inp2==1):
            inp1=0
            
    return inp1,inp2

def or_bt(op,inp1,inp2):
    if(op==0 or op=='d_bar'):
        inp1=0
        inp2=0
    elif(op==1):
        if(inp1==0):
            inp2=1
        elif(inp2==0):
            inp1=1
    return inp1,inp2        
        
```
